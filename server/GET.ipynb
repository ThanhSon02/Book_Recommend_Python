```python
import numpy as np
import pandas as pd
```

```python
users_df = pd.read_csv('Users.csv')
users_df = users_df.dropna(inplace=True)
users_df
```

```python
books_df = pd.read_csv('Books.csv', dtype="string")
ratings_df = pd.read_csv('Ratings.csv')
```

```python
ratings_df
```

```python
ratings_df = ratings_df.drop_duplicates(subset=['User-ID', 'ISBN'], keep='last')
ratings_df
```

```python
ratings_df = ratings_df[ratings_df['Book-Rating'] != 0]
ratings_df
```

```python
books_df = books_df.drop(books_df[books_df["Year-Of-Publication"] == '0'].index)

ratings_mean = ratings_df.groupby("ISBN").agg({'Book-Rating': 'mean', 'User-ID': 'count'}).rename(columns={'User-ID': 'Rating-Count'})
books_df = pd.merge(books_df, ratings_mean, on='ISBN', how='left')
```

```python
popular = books_df.sort_values('Rating-Count', ascending=False)
popular.reset_index(drop=True, inplace=True)
popular.head(10)
```

```python
highestRated = books_df[books_df["Rating-Count"] > 50].sort_values('Book-Rating', ascending=False)
highestRated.reset_index(drop=True, inplace=True)
highestRated.head(10)
```

```python
authors_df = books_df.groupby('Book-Author').agg({'Book-Rating': 'mean', 'ISBN': 'count'}).rename(columns={'ISBN': 'Books-Count'})
authors_rating = authors_df[['Book-Rating', 'Books-Count']]
authors_rating = authors_rating[authors_rating["Books-Count"] > 3]
authors_rating
```
